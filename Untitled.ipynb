```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def cargarGen(path):
    gen = {}
    with open(path, 'r') as archivo:
        tam = int(archivo.readline())
        for first in range(tam):
            linea = archivo.readline()
            linea = list(map(int, linea.split()))
            for snd in range(len(linea)):
                gen[(first, first+snd)] = linea[snd]
    return gen,tam
```

```python
g0, tam = cargarGen("Mediciones/ps20_gr100_ng10/gen0")
```

```python
G = nx.Graph()

pos = {}
size = []
cnt = 0

for k in g0:
    G.add_node(k)
    pos[k] = (cnt // tam, cnt % tam) 
    s = (abs(g0[k] - (k[0] + k[1])))
    if k[0] != k[1]:
        s /= (k[1]-k[0])
    s += 5
    size.append(s)
    cnt += 1
```

```python
'''
Para el dibujo:
    - Cada nodo es un subintervalo [i, j)
    - Para cada nodo el tamao depende de la distancia al centro del intervalo
'''
nx.draw(G, pos=pos, node_size=size)
plt.show()
```
